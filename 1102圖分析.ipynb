```python
import pandas as pd
```

```python
url='https://rate.bot.com.tw/gold/chart/year/TWD'
df=pd.read_html(url)
```

```python
df
```

```python
len(df)
```

```python
df[0]
```

```python
gold=df[0]
```

```python
gold=gold[['日期','本行賣出價格']]
```

```python
gold
```

```python
gold.index=pd.to_datetime(gold['日期'],format="%Y/%m/%d")
```

```python
gold
```

```python
gold.sort_index(inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family']=['Microsoft YaHei']
```

```python
gold["本行賣出價格"].plot(kind="line",figsize=[20,8],xlim=("2020-11-03","2021-11-02"))
```

```python
gold['ma5']=gold["本行賣出價格"].rolling(window=5).mean()
```

```python
gold['ma30']=gold["本行賣出價格"].rolling(window=30).mean()
```

```python
gold[["本行賣出價格","ma5","ma30"]].plot(kind="line",figsize=[20,8],xlim=("2020-11-03","2021-11-02"))
```
