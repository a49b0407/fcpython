```python
import pandas as pd
url='https://od.cdc.gov.tw/eic/Weekly_Age_County_Gender_19CoV.csv'
df=pd.read_csv(url)
```

```python
df
```

```python
df.head(10)#前10筆
```

```python
df.tail()#後5筆
```

```python
#排序索引值遞增排序
df.sort_values("發病週別",ascending=True)
```

```python
df.sort_values(["發病週別","確定病例數"],ascending=False)#遞減false
```

```python
df.sort_values(["發病週別","確定病例數"],ascending=[False,True])#兩個值合併排序
```

```python
df
```

```python
df['縣市']=='台北市'
df.縣市=='台北市'
```

```python
df[df["縣市"]=="台中市"]
```

```python
df[(df["縣市"]=="台中市")&(df["性別"]=="M")]
```

```python
df[(df["縣市"] == "新北市") | (df["縣市"] == "高雄市") | (df["縣市"] == "台中市")]
#一次咧出堆個用 "|管線"
```

```python
df.groupby("縣市")["確定病例數"].sum()#加總
```

```python
df.groupby("縣市")["確定病例數"].max()#最大
```

```python
df.groupby("縣市")["確定病例數"].count()#筆數(回報筆數)
```

```python
df.groupby("縣市")["年齡層"].max()#年齡層縣市
```

```python
df.groupby("縣市")["確定病例數"].agg(["sum","max","min","count","mean"])
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"]=["Microsoft YaHei"]#圖的字型
```

```python
df.groupby("縣市")["確定病例數"].sum().plot(kind="bar",figsize=(16,7))
#bar直方圖,figsize放大
```

```python
#pie圓餅圖line則線圖
df.groupby("縣市")["確定病例數"].sum().plot(kind="pie",figsize=(16,7))
```

```python
#marker節點
df.groupby("發病週別")["確定病例數"].sum().plot(kind="line",figsize=(16,7),marker="s",grid=True)
```
