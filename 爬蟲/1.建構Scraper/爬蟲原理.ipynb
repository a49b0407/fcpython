# <span style="color:red ">何謂爬蟲?
----
##   爬蟲就是一個擷取、複製、過濾及收集資料的過程。
----


#  <span style="color:red"> < HTTP > -網路對話的方式

###  <span style="color:red ">要了解爬蟲，要先了解Http...
----
    我們要用英文溝通，你就需要懂英文文法。
    同樣，你如果要用網路溝通，你就要懂http，這就是client-server之間溝通的方式。
    主要分成:HTTP Requests 和 HTTP Responses
----
   
 ## 請注意
----
    client-server之間，單純就是要文件的關係，要到就關係結束
    所以又稱為stateless
----


###  <span style="color:red ">< Resource >-指的就是server背後的各式各樣資料
----


###  <span style="color:red ">< MIME >(Multipurpose Internet Mail Extensions)-定義了resource的類型
## [多用途網際網路郵件擴展](https://zh.wikipedia.org/wiki/%E5%A4%9A%E7%94%A8%E9%80%94%E4%BA%92%E8%81%AF%E7%B6%B2%E9%83%B5%E4%BB%B6%E6%93%B4%E5%B1%95)
----
    你要怎麼解讀你拿到的resource，你需要MIME的幫忙
----


###  <span style="color:red ">< URLs >(uniform resource locator)-統一資源定位符 定義了resource的位置
----
    1. URL的第一部分稱為scheme，它描述了用於訪問資源的協議。
    2. 第二部分就是網址所在
    3. 第三部分就是你要的資源所在地點
    
----


###  <span style="color:red ">來個更清楚的圖
---


##  <span style="color:red ">< Transactions > 一個完整的request+一個完整的response，我們稱為一個Transactions
----
    整個Transactions的過程，不論request還是response，都會用一個制式化的資料格式來傳遞資料，我們稱為HTTP messages
    
----


### <span style="color:red ">< Request >Client端發出Request給server的這個動作，我們稱為HTTP methods。主要有兩種方法:
----
    1. GET-
       GET 類似明信片一般的寄送方式，你可以把要傳遞的資訊寫在信封( head )，寫到滿為止，價格比較便宜
    2. POST
       POST 就是信封內有裝信件的寄送方式（具有body），不但信封可以寫東西，信封內 (body) 還可以置入你想要寄送的資料或檔案。
----
### [參考連結](https://medium.com/%E5%B0%8F%E5%B0%8F%E8%AA%AA%E6%9B%B8%E4%BA%BA/%E7%B6%B2%E9%A0%81get-%E8%88%87-post-%E5%B7%AE%E7%95%B0-%E7%A7%91%E6%99%AE%E5%A3%B9%E9%BB%9E%E9%80%9A-94cbaa666fdb)

### <span style="color:red ">< status code >Client端發出Request後，會收到Response，在Response message裡面會有 status code:
----
    status code是一個三位數的數字，它代表了response的狀態，有成功獲得資料嗎?
    
----
    status code的意義
    200 (OK, request succeeded)
    404 (Not found; requested resource cannot be found)
    500 (Internal server error)
    204 (No content to be sent)
    401 (Unauthorized request was made to the server)
----


### <span style="color:red ">< 插句話 >你有想過你看網頁的過程，背後到底發生甚麼事嗎?
----
    其實你看到的網頁，是你的瀏覽器去各地把resource都找全了以後，全部最後拼起來，
    給你看的結果。
    
----
 


## <span style="color:red ">< Messages >你可能知道，看網站要輸入網址
### <span style="color:red ">但是你知道，你可以看到網頁，是因為這背後有瀏覽器幫你寫了封信嗎...網址->Message
    
----
### <span style="color:orange ">Messages 就兩種 [Request Message]  [Response Message]  
----
### <span style="color:red ">< Messages 的格式 >   

----
    Start line- 
        如果是request，那就是說一下這個request要做什麼?
        如果是response，那就是說一下request處理後的結果。
    
    Header fields-
        用鍵值對表示，用來描述request和response的相關資料。
    
    Body-
        你就把它當備註欄吧!各種想要傳的訊息都可以寫在這裡。
----


### <span style="color:red ">< 舉個例 >   

----
  1. request一開頭就說了，要用Get method，要拿到tools.html，用的是HTTP 1.0 protocol.
  2. request message 沒有 body 的部分，沒話要說
  3. response 一開頭就告訴你HTTP規格，success status code
  4. body裡面就是你要的html
  5. header裡面的Content-Type，就是這個檔案的MIME
----


### <span style="color:red ">當你發出request 或是收到 response的時候，你都可以收到headers，會有以下幾種分類:
----
    
####  General headers
          Can appear in both request and response messages
####  Request headers
            Provide more information about the request
####  Response headers
            Provide more information about the response
####  Entity headers
            Describe body size and contents, or the resource itself
####  Extension headers
            New headers that are not defined in the specification

#  <span style="color:red"> < HTML >-Hypertext Markup Language 超文本標記語言
   ## [參考連結](https://www.w3schools.com/html/default.asp)
----
    基本上網頁主要就是text, images, style sheets, and scripts所組成。
    其中HTML是最主要的架構。
----
  <範例>
```
    <p>normal paragraph tags</p>
    <h1>heading tags there are also h2, h3, h4, h5, h6</h1>
    <a href="https://www.google.com">Click here for Google.com</a>
    <img src="myphoto1.jpg" width="300" height="300" alt="Picture" />
    <br />
```
      
-----
     1. <p> and <h1> 我們稱為HTML elements或是node
     2. <img>裡面的鍵值對，我們稱為attributes

#  <span style="color:red"> < XML >-Extensible Markup Language (XML)可延伸標記式語言
   ## [參考連結](https://www.w3schools.com/xml/default.asp)
----
    1. 是一種標記式語言。標記指電腦所能理解的資訊符號，通過此種標記，電腦之間可以處理包含各種資訊的文章等。
    2. XML是由相關人士自由決定的標記式語言，這就是語言的可延伸性。
    3. 透過node格式的標記，就可以在電腦間開始Data交換
----
  <範例>
```
    <employees>
        <employee>
            <firstName>Rahul</firstName>
            <lastName>Reddy</lastName>
            <gender>Male</gender>
        </employee>
        
        <employee>
            <firstName>Aasira</firstName>
            <lastName>Chapagain</lastName>
            <gender>Female</gender>
        </employee>
        
        <employee>
            <firstName>Peter</firstName>
            <lastName>Lara</lastName>
            <gender>Male</gender>
        </employee>
    </employees>
```
      
-----
