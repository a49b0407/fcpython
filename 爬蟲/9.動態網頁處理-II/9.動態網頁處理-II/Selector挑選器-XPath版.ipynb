### <span style="color:red ">為什麼我們需要XPath
---
    因為很多元素，我們沒有辦法用ID和Name來定位(因為元素裡面沒有設定啊!!)  
    那我們就需要靠XPath來幫我們解決定位的問題。
---

### <span style="color:red ">XPath表示方式  
    
-----
    xpath =//tag_name[@attribute_name= 'value']
    
-----
    
    tag_name       -元素名  
    @              -標註挑選的屬性名
    attribute_name -屬性名
    value          -屬性值
----

### <span style="color:red ">absolute path-絕對路徑表示法
------
    從根元素一路往下找。
    /這是絕對路徑符號。
    所以會寫成這樣:
    /html/body/div[1]/div/div[1]/div/div[1]/div[2]/div[2]/div/span/b[1]
------
    特點：
    這個路徑是從網頁起始標籤開始一直到要定位的元素的路徑，
    如果要定位的元素在頁面最下面，則這個Xpath路徑會非常長。
    如果在要定位的元素與頁面開始之間的元素有任何增減，元素定位就會失敗。

### <span style="color:red ">relative path-相對路徑表示法
------
    基本上你要先定義一個你開始的元素selected element
    然後用這個開始元素的角度去描述你的需求元素desired element.
    //是相對路徑符號
    會寫成這樣:
    //input

```python
url = 'http://www.webscrapingfordatascience.com/complexjavascript/'



```

### <span style="color:red "><答案>

```python
from selenium import webdriver

url = 'http://www.webscrapingfordatascience.com/complexjavascript/'
driver = webdriver.Chrome()
driver.get(url)

# Use an implicit wait for cases where we don't use an explicit one
driver.implicitly_wait(10)
div_element = driver.find_element_by_xpath('//div[@class="quote"]')
div_element.text
    
```

### <span style="color:red "><小測驗>請用XPath的軸度定位方式幫我把下列網址的前三個quote內容印出來，可以嗎?

```python
url = 'http://www.webscrapingfordatascience.com/complexjavascript/'


```

### <span style="color:red "><答案>

```python
from selenium import webdriver

url = 'http://www.webscrapingfordatascience.com/complexjavascript/'
driver = webdriver.Chrome()
driver.get(url)


driver.implicitly_wait(10)
#注意一下，要找多個，你需要加s
div_element = driver.find_elements_by_xpath('//div[@class="jscroll-added"]/child::div[@class="quote"]')
for i in div_element:
    print(i.text)
```

# <span style="color:red ">XPath Functions
---
    目前網頁日益複雜，單純用id或是name去定位，可能會挑出一堆不相關的元素， 
    而很難去定位出目標元素。
    以及再加上前台畫面會因為Javascript重render的關係，讓元素可能會重新排列。
    所以Xpath推出Xpath functions，讓我們可以更動態的去定位目標元素。


# <span style="color:red ">contains( )
---
    //xpath[contains(@attribute, 'attribute value')] 
---
    1. contains()會出現在[]裡面
    2. 定義了屬性和值
    


### <span style="color:red "><小測驗>請試著用Contains()的定位方式幫我把MOMO的所有品項分類列印出來可以嗎?
---


### <span style="color:red "><答案>

```python
from selenium import webdriver

url = 'https://www.momoshop.com.tw/main/Main.jsp'
driver = webdriver.Chrome()
driver.get(url)
driver.implicitly_wait(10)

#contains允許"部分"符合
div_element = driver.find_elements_by_xpath('//li[contains(@id,"C1")]')
for i in div_element:
    print(i.text)

driver.quit()
```

# <span style="color:red ">text()
    有時候，你想要定位的元素，位置隱密就算了!    
    他還沒有ID、沒有Name、沒有Class，就是一個光溜溜的元素。  
    比如:<button type="button"> 送出 <button/> 
    那還怎麼定位??
    
---
    //xpath[text() = 'login-in'] 
---
    1. text()會出現在[]裡面
    2. 依照帳面文字尋找
    


### <span style="color:red "><小測驗>請幫我從Yahoo首頁進去，點擊"汽車機車"，進去那一頁，可以嗎?

### <span style="color:red "><答案>

```python
from selenium import webdriver

url = 'https://tw.yahoo.com/?p=us'
driver = webdriver.Chrome()
driver.get(url)

#簡單明瞭，不再露露長!!
driver.find_element_by_xpath('//*[text()="汽車機車"]').click()
```

# <span style="color:red ">starts-with( )
    contains()可以讓你找到"部分文字"類似的元素，
    但是它沒有說這些部分文字會出現在哪裡?  
    如果你確定你的目標元素是"開頭幾個字"...   
    請用starts-with( )
    
    
---
    //xpath[starts-with(@attribute, 'attribute value')]



### <span style="color:red "><小測驗>剛剛你已經可以點擊Yahoo的"汽車機車"，進去那一頁，可以幫忙一"汽"呵成嗎?只有一個汽，就找到元素?

### <span style="color:red "><答案>

```python
from selenium import webdriver

url = 'https://tw.yahoo.com/?p=us'
driver = webdriver.Chrome()
driver.get(url)
driver.implicitly_wait(10)

#簡單明瞭，不再露露長!!
driver.find_element_by_xpath('//*[starts-with(text(), "汽")]').click()
```

### <span style="color:red "><小測驗>女王的棋局，很紅的一部美劇，請用XPath的方式幫我把它的歌單爬出來可以嗎?
### [目標網址](https://open.spotify.com/playlist/4NOind9oIIUxaLR8uiWnnd)
------


```python
url = 'https://open.spotify.com/playlist/4NOind9oIIUxaLR8uiWnnd'



```

### <span style="color:red "><答案>

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = 'https://open.spotify.com/playlist/4NOind9oIIUxaLR8uiWnnd'
driver = webdriver.Chrome()
driver.get(url)


# Use an implicit wait for cases where we don't use an explicit one

WebDriverWait(driver, 10).until(
                            EC.presence_of_all_elements_located(
                               (By.XPATH, '//*[@id="main"]/div/div[2]/div[3]/main/div[2]/div[2]/div/div/div[2]/section/div[2]/div[3]/div/div[2]/div[2]/div')
                            )
                         )

div_elements = driver.find_elements_by_xpath('//*[@id="main"]/div/div[2]/div[3]/main/div[2]/div[2]/div/div/div[2]/section/div[2]/div[3]/div/div[2]/div[2]/div')
for i in div_elements:
     target=i.text.split('\n')
     print('歌曲名稱:',target[0])
     print('歌手名稱:',target[1])
     print('專輯名稱:',target[2])
     print('新增日期:',target[3])
     print('歌曲長度:',target[4])
     print('-'*50)

driver.quit()
```
