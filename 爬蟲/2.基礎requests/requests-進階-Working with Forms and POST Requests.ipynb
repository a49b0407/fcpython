# <span style="color:red ">GET的方法特色
-----
    1.這就是一般跟server提出request的方法。
    2.明信片的寄送方法
    2.基本上這樣的請求方法，就是假設你會常常用，用很多次。
    

# <span style="color:red ">POST的方法特色
-----
    1.這是用信封把信包起來的寄信方式，不容窺探。
    2.這種請求方法，是假設你提出請求以後，有一個特定動作已經啟動，server需要時間處理。
    3.你可能需要等待他一個回覆，如果這時候你斷掉了，這個動作就會終止，沒有結果。

````python
#關鍵是網頁裡面有一個隱藏的component，帶有密碼的value
import requests
import re

url = 'http://www.webscrapingfordatascience.com/postform3/'

r = requests.get(url)

pattern=re.compile(r'\svalue="(\w){32}"')
result=pattern.search(r.text)
p_val=result.group()[-33:-1]  

formdata = {
'name': 'Seppe',
'gender': 'M',
'pizza': 'like',
'haircolor': 'brown',
'comments': '',
'protection': p_val
}
r = requests.post(url, data=formdata)
print(r.text)
````

````python
#將來會講到的模組
import requests
from bs4 import BeautifulSoup

url = 'http://www.webscrapingfordatascience.com/postform3/'

r = requests.get(url)

html_soup = BeautifulSoup(r.text, 'html.parser')
p_val = html_soup.find('input', attrs={'name': 'protection'}).get('value')

formdata = {
'name': 'Seppe',
'gender': 'M',
'pizza': 'like',
'haircolor': 'brown',
'comments': '',
'protection': p_val
}
r = requests.post(url, data=formdata)
print(r.text)
````
