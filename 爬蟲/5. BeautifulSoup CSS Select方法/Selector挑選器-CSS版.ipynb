### <span style="color:red ">HTML　＆　XML
-----
    HTML是大家公訂好的node版本
    XML是你可以自訂的node版本
    但是不論是哪一種
    我們都會用樹狀結構圖來分析這些檔案的結構，我們稱為DOM
----

----
    簡單來說，當某個element是包在另一個element裡面的時候，
    我們稱這個是parent-child relationship父子關係。
    而因為這樣的分析結構產生，讓我們有更便利的方式在檔案裡面尋找我們要的資料
----
    而要在檔案裏面遊走，尋找資料，
    我們需要兩個工具
    
    *  XPath 
    *  CSS selectors
