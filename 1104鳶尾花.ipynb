```python
#花蕊資料
from sklearn import datasets
iris=datasets.load_iris()
iris.keys()
```

```python
iris['data']
```

```python
#花資料儲存
import pandas as pd
x=pd.DataFrame(iris['data'],columns=iris['feature_names'])#DataFrame資料表格化
y=pd.DataFrame(iris['target'],columns=['target_name'])
```

```python
iris['feature_names']
```

```python
x
```

```python
y
```

```python
#資料合併
data=pd.concat([x,y],axis=1)#1是水平合併2是垂直
```
